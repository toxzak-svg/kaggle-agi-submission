# TemporalBench — Kaggle AGI Submission
## Measures temporal reasoning: validity windows vs decay functions
**Competition:** Measuring Progress Toward AGI - Cognitive Abilities
**Track:** Reasoning (temporal reasoning)

Tasks:
1. **TemporalBench-v1-AsOfQA** — retrieve what was true at a specific point in time
2. **TemporalBench-v1-ChangeDetection** — detect whether facts changed between two time points
3. **TemporalBench-v1-CausalQuery** — trace causal chains through time
4. **TemporalBench-Reversion** — adversarial reversions (facts that flip back)
5. **TemporalBench-CausalReasoning** — who held a role before someone else took over

**Key finding:** Standard evaluation misses systematic temporal blindness — models fail on recent facts while succeeding on old ones. Validity windows consistently beat decay functions.

```python
# Setup
!pip install -q "protobuf==5.29.6" kaggle-benchmarks
import kaggle_benchmarks as kbench
```

```python
# Import task definitions
from tasks.as_of import (
    temporalbench_v1_asofqa,
    temporalbench_v2_asofqa,
    temporalbench_v3_asofqa,
)
from tasks.change_detection import (
    temporalbench_v1_changedetection,
    temporalbench_v2_changedetection,
)
from tasks.causal_trace import (
    temporalbench_v1_causalquery,
    temporalbench_v2_causalquery,
)
from tasks.reversion import (
    temporalbench_reversion,
    temporalbench_causalreasoning,
    temporalbench_multireversion,
)

print("All tasks loaded successfully.")
```

```python
# Task 1: TemporalBench-v1-AsOfQA
print("Running TemporalBench-v1-AsOfQA...")
result_v1_asof = temporalbench_v1_asofqa.run(kbench.llm)
print(f"v1 AsOfQA accuracy: {result_v1_asof}")
```

```python
# Task 2: TemporalBench-v1-ChangeDetection
print("Running TemporalBench-v1-ChangeDetection...")
result_v1_cd = temporalbench_v1_changedetection.run(kbench.llm)
print(f"v1 ChangeDetection F1: {result_v1_cd}")
```

```python
# Task 3: TemporalBench-v1-CausalQuery
print("Running TemporalBench-v1-CausalQuery...")
result_v1_cq = temporalbench_v1_causalquery.run(kbench.llm)
print(f"v1 CausalQuery accuracy: {result_v1_cq}")
```

```python
# Task 4: TemporalBench-Reversion (adversarial)
print("Running TemporalBench-Reversion...")
result_reversion = temporalbench_reversion.run(kbench.llm)
print(f"Reversion accuracy: {result_reversion}")
```

```python
# Task 5: TemporalBench-CausalReasoning
print("Running TemporalBench-CausalReasoning...")
result_causal = temporalbench_causalreasoning.run(kbench.llm)
print(f"CausalReasoning accuracy: {result_causal}")
```

```python
# Task 6: TemporalBench-MultiReversion
print("Running TemporalBench-MultiReversion...")
result_multirev = temporalbench_multireversion.run(kbench.llm)
print(f"MultiReversion accuracy: {result_multirev}")
```

```python
# Summary
print("\n" + "="*50)
print("TEMPORALBENCH RESULTS SUMMARY")
print("="*50)
print(f"v1-AsOfQA:          {result_v1_asof:.4f}")
print(f"v1-ChangeDetection: {result_v1_cd:.4f}")
print(f"v1-CausalQuery:      {result_v1_cq:.4f}")
print(f"Reversion:           {result_reversion:.4f}")
print(f"CausalReasoning:     {result_causal:.4f}")
print(f"MultiReversion:      {result_multirev:.4f}")
print("="*50)

# Composite TRS score (geometric mean of main metrics)
import math
scores = [result_v1_asof, result_v1_cd, result_v1_cq]
valid = [s for s in scores if s > 0]
if valid:
    trs = math.prod(valid) ** (1/len(valid))
    print(f"Temporal Reasoning Score (TRS): {trs:.4f}")
    print("="*50)
```
